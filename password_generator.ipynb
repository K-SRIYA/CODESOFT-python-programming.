```python
import random
import string
import ipywidgets as widgets
from IPython.display import display, clear_output

# Function to generate password
def generate_password(b):
    clear_output(wait=True)
    display(ui)  # Ensure the UI is displayed again after clearing output

    length = int(length_input.value)
    complexity = complexity_dropdown.value

    characters = string.ascii_letters  # Letters
    if complexity in ['Medium', 'High']:
        characters += string.digits  # Add digits
    if complexity == 'High':
        characters += string.punctuation  # Add punctuation

    password = ''.join(random.choice(characters) for i in range(length))
    password_label.value = f"Generated Password: {password}"

# Creating UI elements
length_input = widgets.IntText(
    value=8,
    description='Length:',
    min=1
)
complexity_dropdown = widgets.Dropdown(
    options=['Low', 'Medium', 'High'],
    description='Complexity:'
)
generate_button = widgets.Button(description="Generate Password")
password_label = widgets.Label(value="Generated Password:")

# Setting up button click event handler
generate_button.on_click(generate_password)

# Displaying the UI
ui = widgets.VBox([length_input, complexity_dropdown, generate_button, password_label])
display(ui)
```
